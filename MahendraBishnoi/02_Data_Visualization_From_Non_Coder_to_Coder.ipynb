## Scatter Plots
4. [Tutorial Link](https://www.kaggle.com/alexisbcook/scatter-plots)

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Load the Data
In this tutorial we will work with (synthetic) dataset of insurance charges to understand why some customers pay more than others

```python
# path of file to be read
path = "data_for_02_Data_Visualization/insurance.csv"

# read file into a variable
insurance_data = pd.read_csv(path)
```

```python
# print first five rows
insurance_data.head()
```

### Create Scatter Plot
To create a simple Scatter Plot, we use `sns.scatterplot` command and specify values for
1. horizontal axis (x-axis)
2. vertical axis (y-axis)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

Scatter Plot shows that `bmi` and `charges` are **positively correlated** because normally people with higer bmi tend to pay more charges. 

To double check we will add **regression line** (line that best fits the data). We can do so by `sns.regplot` command

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

### Color coded Scatter Plots
We can use scatter plot to show relationship between **three** variables. One way to do this is by color coding the points.

To understand how smoking affects the relationship between BMI and insurance costs, we can color code the points by `smoker` and plot `bmi` and `charges` on axes.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

Above plot shows that non-smokers tend to pay much less compared to smokers with increasing bmi.

We can use `sns.lmplot` command to plot two regression lines corresponding to smokers and non-smokers respectively.

```python
plt.figure(figsize=(10, 6))
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

From above plot we can confirm our statement as regression line corresponding to smoker has much higer slope than non-smoker.

### Categorical Scatter Plot
command: `sns.swarmplot` <br/>
Scatter plots are use to show relationship between two continuous variables. Categorical scatter plots feature categorical variable such as `smoker`, `region` etc. on one of the main axes.

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

## Distributions
**Histograms and Density Plots**
5. [Tutorial Link](https://www.kaggle.com/alexisbcook/distributions)

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Load and Examine the Data

```python
# path of file to be read
path = "data_for_02_Data_Visualization/iris.csv"

# Read the file into a variable
iris_data = pd.read_csv(path, index_col="Id")

# print first five rows of the data
iris_data.head()
```

### Histogram
Create a histogram to see how petal length varies in iris flowers. command: `sns.distplot`

```python
sns.distplot(a=iris_data["Petal Length (cm)"], kde=False)

# a = choose the column we'd like to plot
```

### Density Plot
Known as **kernel density estimate (KDE)** plot. Can think of it as smoothened histogram.

command = `sns.kdeplot` and use `shade=True` to color area below the curve.

```python
# KDE plot
sns.kdeplot(data=iris_data["Petal Length (cm)"], shade=True)
```

### 2D KDE Plot
We are not restricted to a single column when creating KDE plot. We can 2D KDE plot with `sns.jointplot` command.

```python
sns.jointplot(x=iris_data["Sepal Width (cm)"], y=iris_data["Petal Width (cm)"], kind="kde")
```

In the above plot the color coding shows us likeliness of seeing different combinations of sepal width and petal width. Darker color means more likeliness.

### Color coded plots

```python
# Paths of files to be read
iris_set_path = "data_for_02_Data_Visualization/iris_setosa.csv"
iris_ver_path = "data_for_02_Data_Visualization/iris_versicolor.csv"
iris_vir_path = "data_for_02_Data_Visualization/iris_virginica.csv"

# Read the files into variables
iris_set_data = pd.read_csv(iris_set_path, index_col="Id")
iris_ver_data = pd.read_csv(iris_ver_path, index_col="Id")
iris_vir_data = pd.read_csv(iris_vir_path, index_col="Id")

# print first five rows of iris_versicolor data
iris_ver_data.head()
```

```python
# set width and height of figure
plt.figure(figsize=(10, 6))

# Histogram for each species
sns.distplot(a=iris_set_data["Petal Length (cm)"], kde=False, label="Iris-setosa")
sns.distplot(a=iris_ver_data["Petal Length (cm)"], kde=False, label="Iris-versicolor")
sns.distplot(a=iris_vir_data["Petal Length (cm)"], kde=False, label="Iris-virginia")

# Add title
plt.title("Histogram of Petal Length (in cm), by species")

# For legend to show
plt.legend()
```

```python
# set width and height of figure
plt.figure(figsize=(10, 6))

# kde plot for each species
sns.kdeplot(data=iris_set_data["Petal Length (cm)"], label="Iris-setosa", shade=True)
sns.kdeplot(data=iris_ver_data["Petal Length (cm)"], label="Iris-versicolor", shade=True)
sns.kdeplot(data=iris_vir_data["Petal Length (cm)"], label="Iris-virginia", shade=True)

# Add title
plt.title("KDE of Petal Length (in cm), by species")
```

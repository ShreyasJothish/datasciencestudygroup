## Line Charts
2. [Tutorial link](https://www.kaggle.com/alexisbcook/line-charts)

```python
# First of all we will setup our notebook for plotting
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

The Dataset for this tutorial (Line Charts) tracks global daily streams on music streaming service Spotify. We will focus on five popular songs from 2017-18. <br/>
1. "Shape of You", by Ed Sheeran
2. "Despacito", by Luis Fonzi
3. "Something Just Like This", by The Chainsmokers and Coldplay
4. "HUMBLE.", by Kendrick Lamar 
5. "Unforgettable", by French Montana

### Load the Data

```python
# specify the path to dataset
path = "data_for_02_Data_Visualization/spotify.csv"

# Load the file 'spotify.csv' into a variable
spotify_data = pd.read_csv(path, index_col = "Date", parse_dates = True)
```

 ### Examine the Data

```python
# print first five rows of the data
spotify_data.head()
```

Here we can see that first date is 6th January 2017 and Shape of You was streamed `12287078` times on 6th Jan. <br/>
We are seeing `NaN` for songs other than `Shape of You` because they released later. <br/>
> Empty entries appear as `NaN` which is short for Not a "Number"

```python
# print last five rows of the data
spotify_data.tail()
```

### Plot the Data

```python
# Line chart showing daily global stream of each song
sns.lineplot(data = spotify_data)
```

```python
# We can specify some additional values such as size of figure

# Specify figure size
plt.figure(figsize=(17, 6))

# Add a title to the figure
plt.title("Daily streams of popular songs in 2017-18")

# Create line chart
sns.lineplot(data = spotify_data)
```

### Plot a subset of the Data
We have learned how to plot all the columns of data. Now we will learn how to plot selected columns from the data.

```python
# create a list of all columns 
list(spotify_data.columns)
```

```python
# Now we will plot first two columns

# set figure size 
plt.figure(figsize=(17, 6))

# set title
plt.title("Daily streams of popular songs in 2017-18")

# plot first column "Shape of You"
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# plot second column "Despacito"
sns.lineplot(data=spotify_data["Despacito"], label="Despacito")

# set xlabel
plt.xlabel("Date")
```

## Bar Charts and Heat Map
3. [Tutorial Link](https://www.kaggle.com/alexisbcook/bar-charts-and-heatmaps)

### Setup Notebook

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

In this tutorial we will work with a dataset from US Department of Transportation that tracks flight delays.

### Load the Data

```python
# specify path of dataset
path = "data_for_02_Data_Visualization/flight_delays.csv"

# read data from file into a variable
flight_data = pd.read_csv(path, index_col="Month")
```

### Examine the Data

```python
flight_data
```

### Bar Chart
Create a bar chart for "Spirit Airlines" (Code: "NK") flight dealys by month

```python
# set figure size
plt.figure(figsize=(12, 6))

# set tilte
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Create Bar Chart using seaborn
sns.barplot(x=flight_data.index, y=flight_data.NK)

# Add label for vertical axis
plt.ylabel("average arrival delay (in minutes)")
```

### Heat Map

```python
# set width and height of figure
plt.figure(figsize=(15, 6))

# Add title
plt.title("Average Arrival Delay for each Airline, by Month")

# Create heatmap showing average arrival delay 
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis 
plt.xlabel("Airline")
```
